```python
use = pd.read_excel("harvey-usage-start_2024-05-01_end_2024-12-30.xlsx")
```

```python
use['User'] == use['User'].str.strip().str.lower()
use['User'][:10]
```

```python
filtered_use = use[use['User'].isin (not_in_firm['Email'])]
```

```python
len(filtered_use)
```

```python
filtered_use
```

```python
user_stats = (
    filtered_use.groupby('User')
    .agg(Queries=('User', 'size'), MostRecentQuery=('Time', 'max'))
    .reset_index()
)

user_stats = user_stats.sort_values(by='MostRecentQuery', ascending=False)
```
